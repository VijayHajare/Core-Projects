```python
import numpy as np
```

```python
R0 = 1
r0 = 1.2
R_limit = np.linspace(0.9,1,100)
```

```python
p_all = []
for R in R_limit:
  n = 10000
  h = (R0-R)/n
  #h =round(h,3)
  R_all = []
  i = 0
  for i in range(n):
    t = R+h*i
    R_all.append(t)
    i= i+1
  y = [] #value of intergation function for different value of R in a given range of R0 to R
  for R in R_all:
    fun = (1/((R**3-R0**3+r0**3)**(1/3)))*((R**6)/(R**3-R0**3+r0**3)) - (((R**3-R0**3+r0**3)**(1/3))*((R**2)/((R**3-R0**3+r0**3)**(2/3)))/R**2) #function of integration
    y.append(fun)
  odd = range(1, n+1, 2)
  even = range(2,n-1,2)
  odd_list = []
  for i in odd:
    collect = y[i]
    odd_list.append(collect)
  odd_sum = np.sum(odd_list)
  even_list = []
  for i in even:
    collect = y[i]
    even_list.append(collect)
  even_sum = np.sum(even_list)
  total = h*(y[0]+ 4*odd_sum+2*even_sum)
  p = 0.838*(((R**2/((R**3-R0**3+r0**3)**(2/3))))**2) - total
  p_all.append(p)



```

```python
import matplotlib.pyplot as plt
xpoints = R_limit
ypoints = p_all

plt.plot(xpoints, ypoints)
plt.xlabel("R ")
plt.ylabel("p")
plt.title("R vs P ")
plt.grid()
plt.show()
```

```python
srr_all = []
for p,R in zip(p_all,R_limit):
  srr = -p + 0.838*(((R**2)/((R**3-R0**3+r0**3)**(2/3)))**2)
  srr_all.append(srr)
```

```python
xpoints = R_limit
ypoints = srr_all

plt.plot(xpoints, ypoints)
plt.xlabel("R ")
plt.ylabel("Stress in Radial Direction")
plt.title("Stress in Radial Direction vs R ")
plt.grid()
plt.show()
```

```python
r_all = []
for R in R_limit:
  r = (R**3-R0**3+r0**3)**(1/3)
  r_all.append(r)
```

```python
xpoints = R_limit
ypoints = r_all

plt.plot(xpoints, ypoints, color = "green")
plt.xlabel("R ")
plt.ylabel(" r")
plt.title("r vs R ")
plt.grid()
plt.show()
```
